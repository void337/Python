1. Даны три целых числа. Определите, сколько среди них совпадающих. Функция должна вернуть одно из чисел: 3 (если все совпадают), 2 (если два совпадает) или 0 (если все числа различны).

```python
def sovpadenie(nums):
  if(len(nums) == len(set(nums))): return 0
  return  len(nums) - len(set(nums)) + 1
```

```python
nums = [int(i) for i in input("Введите числа в одну строку\n").split()]
print(sovpadenie(nums))
```

2. По данному натуральному n ≤ 9 выведите лесенку из n ступенек, i-я ступенька состоит из чисел от 1 до i без пробелов.

```python
def stairs(n):
  for i in range(n):
      print(''.join(map(str, range(1, i+2))))
```

```python
n = int(input("Введите число n ≤ 9\n"))
while n > 9:
  n = int(input("Введите число n ≤ 9\n"))
stairs(n)
```

3. Дополните код из предыдущего задания, чтобы теперь получалась пирамида. То есть каждая ступень состоит из чисел от 1 до i и обратно.

```python
def piramide(n):
  for i in range(1, n+1):
    s = ''.join(map(str, range(1, i+1))) + ''.join(map(str, range(i-1, 0, -1)))
    print (' ' * (n - i), s)
```

```python
n = int(input("Введите число\n"))
piramide(n)
```

4. Задан список с числами. Напишите функцию, которая добавляет все элементы списка с четными индексами в новый список и возвращает его.

```python
def even(nums):
  return [nums[i] for i in range(len(nums)) if i%2==0]
```

```python
nums = [int(i) for i in input("Введите числа в одну строку\n").split()]
print(even(nums))
```

5. Задан список с числами. Напишите функцию, которая меняет местами наибольший и наименьший элемент и возвращает новый список.

```python
def change(nums):
  max_index = nums.index(max(nums))
  min_index = nums.index(min(nums))
  nums[max_index], nums[min_index] = nums[min_index], nums[max_index]
  return nums
```

```python
nums = [int(i) for i in input("Введите числа в одну строку\n").split()]
print(change(nums))
```

6. Пароль считается надежным, если его длина составляет не менее 12 символов, при этом он должен содержать хотя бы одну заглавную букву, хотя бы одну строчную букву, хотя бы одну цифру и хотя бы один спецсимвол. Напишите функцию, которая обрабатывает строковые данные и возвращает 1 - если они могут быть паролем, и 0 в обратном случае

```python
def check(password):
  sym = "!#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~"
  if len(password) >= 12 and \
     not password.islower() and \
     not password.isupper() and \
     any(c.isdigit() for c in password) and \
     set(password).intersection(set(sym)):
    return 1
  return 0
```

```python
password = str(input("Введите пароль\n"))
check(password)
```

7. Напишите функцию, которая будет производить поиск по значению и выдавать ключ.

```python
def find(dict, value):
  for key, val in dict.items():
    if val == value: return key
  return None
```

```python
my_dict = {"Город": "Ростов", "Страна": "Россия", "Имя": "Кристина"}
value = str(input("Введите значение\n"))
print(find(my_dict, value))
```
