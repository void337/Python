```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 50, 50)
np.random.seed(1)
y1 = np.random.randint(0, 20, 50)
y2 = np.random.randint(10, 30, 50)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

ax1.plot(x, y1, label="Линия 1")
ax2.plot(x, y2, label="Линия 2")
lines_labels = [ax1.get_legend_handles_labels(), ax2.get_legend_handles_labels()]
lines, labels = [sum(l, []) for l in zip(*lines_labels)]
fig.legend(lines, labels, loc='upper left', fontsize=12,  ncol=2, frameon=True, framealpha=1, edgecolor='red', bbox_to_anchor=(1, 0.5))

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, y1, label="Линия 1")
ax.plot(x, y2, label="Линия 2")

ax.legend(loc=0, fontsize=12, frameon=False, labelspacing=2,)

plt.tight_layout()
plt.show()
```

# 2. Линейная диаграмма

```python
plt.axhline(y=0.2, color='red', linestyle='-', linewidth=1)
plt.axhline(y=0.4, color='blue', linestyle='--', linewidth=2)
plt.axhline(y=0.6, color='green', linestyle='-.', linewidth=3)
plt.axhline(y=0.8, color='purple', linestyle=':', linewidth=4)
plt.axvline(x=0.2, color='red', linestyle='-', linewidth=1)
plt.axvline(x=0.4, color='blue', linestyle='--', linewidth=2)
plt.axvline(x=0.6, color='green', linestyle='-.', linewidth=12, alpha=0.5)

plt.fill_between(np.linspace(0, 1, 100), 0.8, 0.6, color='lightblue', alpha=0.5)
plt.fill_between(np.linspace(0, 1, 100), 0.4, color='lightcoral', alpha=0.5)

plt.title("Несколько горизонтальных линий")
plt.show()
```

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)
fig1, ax1 = plt.subplots()
wedges, texts, autotexts = ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90, wedgeprops={'edgecolor':'black', 'linewidth':1})

ax1.axis('equal')
plt.title('Круговая диаграмма')
plt.show()
```

```python
import random

x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y, alpha=0.5, color='red')
ax.set_title('Заголовок', fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=14)
ax.set_facecolor('lightblue')
plt.show()

fig, ax = plt.subplots(figsize=(12, 6))
r, g, b = random.random(), random.random(), random.random()
ax.plot(x, y, color=(r, g, b), label='sin(x)')
ax.text(5, 0.5, 'Текст внутри графика', fontsize=12)
ax.legend(frameon=True, framealpha=1, edgecolor='red', facecolor='lightyellow')
ax.tick_params(axis='x', which='major', labelsize=16)
ax.tick_params(axis='y', which='major', labelsize=12)
plt.show()

fig, ax = plt.subplots()
im = ax.imshow(np.random.rand(10,10))
cbar = fig.colorbar(im, shrink=0.5)
plt.show()

#Скрытие меток текста оси
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xticklabels([])
ax.set_yticklabels([])
plt.show()


#Настройка положения меток осей
fig, ax = plt.subplots()
ax.plot(x, y)
ax.tick_params(axis='x', which='major', pad=15)
plt.show()


#Скрытие оси, границ и пробелов
fig, ax = plt.subplots()
ax.plot(x, y)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.set_xticks([])
ax.set_yticks([])
plt.show()
```
